```python
#Importing required packages
import pandas as pd
```

```python
#Importing the file
df = pd.read_csv("Retail Sales Analysis_raw data.csv")
df.head(5)
```

```python
#remove white spaces
df.columns = df.columns.str.strip()
```

```python
#print(df['sale_date'].unique())
df['sale_date'] = pd.to_datetime(df['sale_date'], format='%d-%m-%Y', errors='coerce')
df['sale_date']
```

```python
df.rename(columns={'quantiy': 'quantity'}, inplace=True)
```

```python
df = df.drop_duplicates() #To remove duplicate rows. if availble
df
```

```python
df.isnull().sum()
```

```python
#Fill missing age values with median age
df['age'] = df['age'].fillna(df.groupby(['gender', 'category'])['age'].transform('median'))
```

```python
#other missing value is replace with overall median value
x=df['age'].median()
df['age'] = df['age'].fillna(x)
```

```python
df['age'].isnull().sum()
```

```python
#Filling the null values by apllying the median
for col in ['total_sale', 'quantity', 'cogs', 'price_per_unit']:
    df[col] = df[col].fillna(df[col].median())
```

```python
#Null values were handeled efficiently
df.isnull().sum() 
```

```python
df.to_csv("Sales_insights_updated.csv", index= False)
```
